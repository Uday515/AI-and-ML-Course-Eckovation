## Association Rule Mining: Market Basket Analysis

### Question:
### Apriori is a statistical algorithm for implementing associate rule mining, that primarily relies on three components: Life, Support and Confidence. Using this algorithm try to find the rules that describe the relation between each of the products that were brought by the customers as described in store_data.csv.

## Import Libraries 

```python
import pandas as pd
import numpy as np 
import csv
```

```python
words=[]
with open("store_data.csv", mode='r') as data:  #loading the csv data
    raw_data = csv.reader(data, delimiter=',')
    for row in raw_data:
        words.extend(row)
        
```

```python
words=list(set(words)) # finding target value 
len(words)
```

```python
words
```

```python
df=pd.read_csv("store_data.csv", delimiter=',',header= None)
df
```

## Creating own table in a usable format 


```python
X={}
for i in words:
    X[i]=[0]*7501
df1=pd.DataFrame(X)
df1    
```

```python
for i in range(7501):
    for j in range(20):
        e =  df.iloc[i,j]
        df1.iloc[i][e] = 1
```

```python
df1.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
freq_itemsets = apriori(df1, min_support=0.01,use_colnames = True)
```

## Pattern by which items where sold 

```python
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, metric='confidence',min_threshold=0.46)
rules.head(10)
```

```python
rules[ (rules['lift'] > 1.2) & (rules['confidence'] > 0.46)]
```
